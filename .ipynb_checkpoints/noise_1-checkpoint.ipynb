```python
import numpy as np
import h5py

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = h5py.File('errormodel/final-errorscan-qf_2-64-4_3-3000.h5', 'r')

p = np.array(f['p'])
qfe = np.array(f['qfe'])
qfp = np.array(f['qfp'])

qfe_m = np.mean(qfe, axis=0)
qfp_m = np.mean(qfp, axis=0)

qfe_s = np.std(qfe, axis=0)
qfp_s = np.std(qfp, axis=0)

print(qfe_m)
print(qfe_s)
```

```python
plt.plot(p, qfe_m, marker='.', markersize=12)
plt.ylim(bottom=0)
```

```python

fig, ax = plt.subplots()
ax.plot([-1,1],[2,0], linestyle='--', color='0.7', linewidth=4, label='Exact')
ax.errorbar(p, qfp_m, yerr=qfp_s, marker='.', markersize=12, linewidth=2, color='tomato',label='AQT')

ax.set_ybound(0.5,1.08)
ax.set_xbound(-0.005, 0.305)

ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)
ax.set_xlabel('$p$',size=14)
ax.set_ylabel('$F_Q$',size=14)
ax.legend(loc='lower left', fontsize=14)

# fig.savefig('figures/fig2b.pdf', dpi=300, frameon=True,bbox_inches="tight")
```
