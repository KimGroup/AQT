# MLE implementation from Qiskit

```python
from qiskit.ignis.verification.tomography import state_tomography_circuits, StateTomographyFitter

Nq = 3

qr = QuantumRegister(Nq)
circuit = QuantumCircuit(qr)

circuit.h(0)
for idx in range(Nq-1):
    circuit.cx(idx, idx+1)
    

qst_ghz = state_tomography_circuits(circuit, qr)


tomo_ghz = StateTomographyFitter(result, qst_ghz)
dm_ibmq = tomo_ghz.fit()

```

```python
from povm import *
from qiskit.quantum_info import state_fidelity

Ne = 4
rhoKB = np.zeros((Ne, 2*Nq), dtype=int)
rhoAmp = np.zeros((Ne), dtype=complex)

# Pure GHZ state
rhoKB[1, :] = 1
rhoKB[2, :Nq] = 1
rhoKB[3, Nq:] = 1

rhoAmp[0] = 0.5
rhoAmp[1] = 0.5
rhoAmp[2] = 0.5
rhoAmp[3] = np.conj(rhoAmp[2])

density = {}
density['KB'] = rhoKB
density['Amp'] = rhoAmp

dm_ghz = Struct2DM(density)

print(state_fidelity(dm_ibmq, dm_ghz))
```

```python
fig_model = plotDM(dm_ibmq, Nq)

# np.save('dm/dm_ibmq_Nq3.npy', dm_ibmq)
# fig_model.savefig('figures/ibmq-MLE.pdf', dpi=300, frameon=True,bbox_inches="tight")
```

```python
Nq = 3

Nset = 3**Nq

collateddata = np.zeros((2700,3), dtype=int)
di = 0
for nset in range(Nset):
    setting = result.results[nset].header.name.split('\'')[1::2]
    
    setstr = ''
    for nq in range(Nq):
        s = setting[nq]
        if s == 'X':
            setstr += '1'
        elif s == 'Y':
            setstr += '2'
        elif s == 'Z':
            setstr += '0'
        else:
            print('Unknown setting: %s'%s)
    
    rawdata = result.get_counts(nset)
    
    print(setstr)
    for st in rawdata:
        # Must be careful with indexing due to IBMQ's indexing convention
        outcome = np.array([3*int(st[2-i])+int(setstr[i]) for i in range(3)])
        
        collateddata[di:di+rawdata[st]] = outcome
        di += rawdata[st]


assert(di==2700)

fname = 'data/ibmq-ghz-Nq3.npy'
# np.save(fname,collateddata)
```

# My own MLE implementation

```python

Na = 6
Ns = len(collateddata)
ptab = np.zeros((Na**Nq), dtype=float)

for ns in range(Ns):
    o = basestr2int(collateddata[ns], Na)
    ptab[o] += 1./Ns
```

```python
from MLE import *
from povm import *

dm_mle = MLE_DM(ptab, Nq, POVM('pauli6'))
fig_model = plotDM(dm_mle, Nq)
plt.show(fig_model)
```

```python
from qiskit.quantum_info import state_fidelity

print(state_fidelity(dm_mle, dm_ibmq))
```
