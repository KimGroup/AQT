```python
import numpy as np
import h5py

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = h5py.File('data/final-epochscan-val_2-64-4_20-8000.h5', 'r')

Nep = np.array(f['Nep'])
cf = np.array(f['cf'])

plt.plot(Nep, cf, marker='.', markersize=12)
plt.ylim(top=1)

print(max(cf))
```

```python
att_q = np.array([10,20,40,60, 80])
att_s = np.array([6000, 8000, 10000, 11500,13000])


fig, ax = plt.subplots()

ax.plot(att_q, att_s, marker='.', markersize=12)

ax.set_ybound(lower=0)
```

```python
fig, ax = plt.subplots()

dataT=np.loadtxt("for_peter/linear_Pauli_0.99.txt")

ax.errorbar(np.log(dataT[:,0]),np.log(dataT[:,1]),dataT[:,3]/dataT[:,1], marker='o', label='RNN',ls='-')


ax.plot(np.log(att_q), np.log(att_s), marker='.', markersize=12,color='tomato', label='AQT')


xticks = np.array([10,20, 40, 80])
yticks = np.array([3000,9000,27000,81000])

xticklabels = ['%i'%x for x in xticks]
yticklabels = ['%i'%y for y in yticks]
ax.set_xticks(np.log(xticks))
ax.set_xticklabels(xticklabels,size=14)

ax.set_yticks(np.log(yticks))
ax.set_yticklabels(yticklabels,size=14)


handles, labels = ax.get_legend_handles_labels()

ax.legend([handles[1],handles[0]], [labels[1],labels[0]], fontsize=14)
ax.set_xlabel('$N_q$',size=14)
ax.set_ylabel('$N_s^*$',size=14)
ax.set_ybound(np.log(3000), np.log(200000))


fig.savefig('figures/fig2a.pdf', dpi=300, frameon=True,bbox_inches="tight")
```
